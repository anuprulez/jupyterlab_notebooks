## Plotting with Seaborn

```python
## https://seaborn.pydata.org/tutorial.html

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Plots for visualising data distribution

```python
## Univariate histograms

penguins = sns.load_dataset("penguins")
sns.displot(penguins, x="flipper_length_mm")
```

```python
sns.displot(penguins, x="flipper_length_mm", hue="species")
```

```python
sns.displot(penguins, x="flipper_length_mm", hue="sex", multiple="dodge")
```

```python
## Kernel density estimation

sns.displot(penguins, x="flipper_length_mm", hue="species", kind="kde", multiple="stack")
```

```python
## Bivariate distributions

sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", hue="species")
```

### Plots for regression models

```python
sns.set_theme(color_codes=True)
tips = sns.load_dataset("tips")
```

```python
sns.regplot(x="total_bill", y="tip", data=tips);
```

```python
sns.lmplot(x="size", y="tip", data=tips, x_estimator=np.mean);
```

```python
## Fit different models

anscombe = sns.load_dataset("anscombe")

sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'I'"),
           ci=None, scatter_kws={"s": 80});
```

```python
tips["big_tip"] = (tips.tip / tips.total_bill) > .15
sns.lmplot(x="total_bill", y="big_tip", data=tips,
           y_jitter=.03);
```

```python
## Conditioned on different variable
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips);
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", data=tips);
```

### Plots for categorical data

```python
tips = sns.load_dataset("tips")
sns.catplot(x="day", y="total_bill", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", hue="sex", kind="swarm", data=tips)
```

```python
sns.catplot(x="total_bill", y="day", hue="time", kind="swarm", data=tips)
```

### Boxplots

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker", kind="box", data=tips)
```

```python
sns.catplot(x="total_bill", y="day", hue="sex", kind="violin", data=tips)
```

```python
sns.catplot(x="total_bill", y="day", hue="sex", kind="violin", bw=.15, cut=0, data=tips)
```
